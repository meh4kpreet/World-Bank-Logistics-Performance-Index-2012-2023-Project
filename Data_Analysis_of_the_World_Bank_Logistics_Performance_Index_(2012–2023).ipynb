Pre vs Post COVID-19 Pandemic Comparisons

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pre_covid = df[df['year'] < 2020]
post_covid = df[df['year'] >= 2020]

cols = ['lpi_score', 'customs_score', 'infrastructure_score',
        'international_shipments_score', 'logistics_competence_and_quality_score',
        'timeliness_score', 'tracking_and_tracing_score']
pre_avg = pre_covid[cols].mean().rename('Pre-COVID (2012‚Äì2019)')
post_avg = post_covid[cols].mean().rename('Post-COVID (2020‚Äì2023)')

comparison_df = pd.concat([pre_avg, post_avg], axis=1)
comparison_df['Absolute Change'] = comparison_df['Post-COVID (2020‚Äì2023)'] - comparison_df['Pre-COVID (2012‚Äì2019)']
comparison_df['% Change'] = (comparison_df['Absolute Change'] / comparison_df['Pre-COVID (2012‚Äì2019)']) * 100

print(" Pre vs Post-COVID Descriptive Comparison:\n")
print(comparison_df.round(3))
comparison_df[['Pre-COVID (2012‚Äì2019)', 'Post-COVID (2020‚Äì2023)']].plot(kind='bar', figsize=(10, 5))
plt.title("Bar Chart: LPI Component Averages ‚Äì Pre vs Post COVID")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## **Top Key Insights from Global LPI Data (2012‚Äì2023)**
1. Global Logistics Performance Peaked Around 2016
Average global LPI score peaked at ~3.0 in 2016, followed by a decline till 2020 and only partial recovery by 2023.

*  Indicates pre-COVID maturity, then pandemic-related disruption.

---


2. COVID-19 Severely Impacted Logistics Efficiency
All LPI components dropped significantly in 2020.

*   Timeliness Score saw one of the sharpest drops.
*   Tracking & Tracing and International Shipments also fell.


---


3. Germany and Singapore Consistently Top Performers
Held #1 and #2 ranks across most years.Both showed minor declines post-COVID but remained global logistics benchmarks.

---


4. India and China Showed Steady Improvement
India‚Äôs LPI score increased from ~3.08 in 2012 to ~3.4 in 2023.


---


5. U.S. Logistics Rank Volatile Despite High Score
United States maintained a good LPI score (~3.5+) but dropped in global ranking post-2018.



---

6. Among all LPI metrics, Timeliness Score shows least volatility over the decade.

üìä Pre vs Post-COVID Key Insights (2012‚Äì2023)
1. Overall Improvement in LPI Score Post-COVID:
Average LPI Score increased from 2.983 (2012‚Äì2019) to 3.046 (2020‚Äì2023), a 2.1% rise, showing slight recovery and adaptation in global logistics post-pandemic.

2. Infrastructure Score Improved Notably:

3. Logistics Competence & Quality saw highest % improvement:

4. Tracking and Tracing Efficiency also grew:
Increased by 3.5%, from 2.996 to 3.103‚Äîhighlighting enhancements in visibility and digital monitoring tools in global supply chains.

5. Customs Procedures Became Slightly More Efficient:
Improved from 2.794 to 2.855 (+2.17%), indicating modest gains in clearance processes.
